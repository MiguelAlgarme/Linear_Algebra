```python
import random
import numpy as np


#1. Develop a python function from scratch that will find the determinants of any x matrix.

def determinant(matrix):
    assert len(matrix) == 0 or (all(len(row) == len(matrix) for row in matrix) and len(matrix) == len(matrix[0])),
    size = len(matrix)
    
    if size == 0:
        return 1.0
    elif size == 1:
        return matrix[0][0]
    elif size == 2:
        return matrix[0][0] * matrix[1][1] - matrix[0][1] * matrix[1][0] 
    
    det = 0.0
    
    for col, cofactor in enumerate(matrix[0]):
        minor = [row[1:] for row in matrix[1:]]  # Removed unnecessary list comprehension
        det += (-1) ** col * cofactor * determinant(minor)

    return det

#2. Develop a python function from scratch that will find both the eigenvectors and eigenvalues of any x matrix.

def eigen_decomposition(matrix, max_iterations=50000):
    assert len(matrix) == 0 or (all(isinstance(row, list) for row in matrix) and len(matrix) == len(matrix[0])), "Input must be a square matrix"

    size = len(matrix)
    if size == 0:
        return [], []
    if size == 1:
        return [matrix[0][0]], [[1.]]

    eigenvalues = []
    eigenvectors = []

    for _ in range(size):
        guess_eigenvector = [random.random() for _ in range(size)]
        norm_guess_eigenvector = sum(a ** 2 for a in guess_eigenvector) ** 0.5
        guess_eigenvector = [a / norm_guess_eigenvector for a in guess_eigenvector]
        

        for _ in range(max_iterations):
            transformed_eigenvector = [sum(a * b for a, b in zip(row, guess_eigenvector)) for row in matrix]

            norm_transformed = sum(a ** 2 for a in transformed_eigenvector) ** 0.5
            guess_eigenvector = [a / norm_transformed for a in transformed_eigenvector]

        eigenvalue = sum(a * b for a, b in zip(transformed_eigenvector, guess_eigenvector))
        eigenvalues.append(eigenvalue)
        eigenvectors.append(guess_eigenvector)


        eigenvalue_outer_product = [[a * b for b in guess_eigenvector] for a in transformed_eigenvector]
        matrix = [[matrix[i][j] - eigenvalue_outer_product[i][j] for j in range(size)] for i in range(size)]

    return eigenvalues, [list(row) for row in zip(*eigenvectors)]



#3. Test your functions from a randomly generated  x matrix.
```
